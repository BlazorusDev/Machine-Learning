```python
def convertRange(x):
    temp = x.split('-')
    if len(temp) == 2:
        return (float(temp[0])+float(temp[1]))/2
    try:
        return float(x)
    except:
        return None
```

```python
data['total_sqft'] = data['total_sqft'].apply(convertRange)
```

```python
def remove_outliers_sqft(df):
    df_out = pd.DataFrame()
    for key, subdf in df.groupby('location'):
        m = np.mean(subdf.price_per_sqft)
        st = np.std(subdf.price_per_sqft)
        reduced_df = subdf[(subdf.price_per_sqft>(m-st)) & (subdf.price_per_sqft<=(m+st))]
        df_out = pd.concat([df_out,reduced_df],ignore_index=True)
    return df_out
data = remove_outliers_sqft(data)
data.shape
```

```python
data.describe()
```

```python
def bhk_outlier_remover(df):
    exclude_indices = np.array([])
    for location, location_df in df.groupby('location'):
        bhk_stats = {}
        for bhk, bhk_df in location_df.groupby('bhk'):
            bhk_stats[bhk] = {
                'mean': np.mean(bhk_df.price_per_sqft),
                'std': np.std(bhk_df.price_per_sqft),
                'count': bhk_df.shape[0]
            }
        for bhk, bhk_df in location_df.groupby('bhk'):
            stats = bhk_stats.get(bhk-1)
            if stats and stats['count']>5:
                exclude_indices = np.append(exclude_indices, bhk_df[bhk_df.price_per_sqft<(stats['mean'])].index.values)
    return df.drop(exclude_indices,axis='index')
data = bhk_outlier_remover(data)
data.shape
```

```python
X = data.drop(columns=['price'])
y=data['price']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,Lasso,Ridge
from sklearn.preprocessing import OneHotEncoder,StandardScaler
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
print("Training set shape:", X_train.shape, y_train.shape)
print("Test set shape:", X_test.shape, y_test.shape)
```

```python
column_trans = make_column_transformer((OneHotEncoder(sparse_output=False),['location']),
                                      remainder='passthrough')
```

```python
# Linaer regression
scalar = StandardScaler()
```

```python
lr = LinearRegression()
```

```python
pipe = make_pipeline(column_trans,scalar,lr)
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred_lr = pipe.predict(X_test)
```

```python
r2_score(y_test,y_pred_lr)
```

```python
# Lasso
lasso = Lasso()
```

```python
pipe = make_pipeline(column_trans,scalar,lasso)
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred_lr = pipe.predict(X_test)
r2_score(y_test,y_pred_lr)
```

```python
# Ridge
ridge = Ridge()
```

```python
pipe = make_pipeline(column_trans,scalar,ridge)
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred_lr = pipe.predict(X_test)
r2_score(y_test,y_pred_lr)
```

```python
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, VotingRegressor

pipe=make_pipeline(column_trans,scalar,RandomForestRegressor(n_estimators=250))
pipe.fit(X_train,y_train)
y_pred=pipe.predict(X_test)
print('rfr',r2_score(y_test,y_pred))
```

```python
estimate=[
    ('lr',lr),
    ('rf',RandomForestRegressor()),
    ('rd',Ridge())
]
```

```python
from sklearn.ensemble import VotingRegressor 
pipe=make_pipeline(column_trans,scalar,VotingRegressor(estimators=estimate))
pipe.fit(X_train,y_train)
y_pred=pipe.predict(X_test)
print('vr',r2_score(y_test,y_pred))
```

```python
from sklearn.model_selection import cross_val_score
cv=cross_val_score(pipe,X_train,y_train,n_jobs=1,cv=10)
```
